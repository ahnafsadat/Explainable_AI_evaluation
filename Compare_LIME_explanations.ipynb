```python
from PIL import Image

def change_size_for_model(image):
    image = Image.fromarray(image)
    resized_image = image.resize((224, 224))
    resized_array = np.array(resized_image)
    return resized_array.astype(np.uint8)

def change_size_for_original(image, original_shape):
    image = Image.fromarray(image)
    resized_image = image.resize(original_shape) #resize image
    resized_array = np.array(resized_image)
    return resized_array.astype(np.uint8)
```

```python
compare_IOU_explanation_list = list()
compare_RMSE_explanation_list = list()

for j in range(len(list_of_dir)):
# for j in range(5):

  try:

    ground_truth_files_path = ground_truth_path +'/'+str(list_of_dir[j])+'/'

    list_of_names = os.listdir(ground_truth_files_path)
    list_of_names.sort()

    ground_truth_explanation_array = list()

    for img_name in list_of_names:
      # Load the image
      image = Image.open(ground_truth_files_path + img_name)
      image = np.array(image)
      image = image[:,:,1]

      # Convert the image to a numpy array
      ground_truth_explanation_array.append(np.array(image))


    original_files_path = original_GRADCAM_explanation_path +'/'+str(list_of_dir[j])+'/'

    original_explanation_array = list()

    for img_name in list_of_names:
      # Load the image
      image = Image.open(original_files_path + img_name)
      image = np.array(image)
      image = image[:,:,1]

      # Convert the image to a numpy array
      original_explanation_array.append(np.array(image))


    for i in range(len(list_of_names)):
      # Load the image

      image_a = ground_truth_explanation_array[i]
      image_b = original_explanation_array[i]

      image_a_bool = (image_a == 255)
      image_b_bool = (image_b == 255)

      # Calculate Intersection and Union
      intersection = np.logical_and(image_a_bool, image_b_bool)
      union = np.logical_or(image_a_bool, image_b)

      # Compute the IoU
      iou = np.sum(intersection) / np.sum(union)

      compare_IOU_explanation_list.extend([iou])

      a_r, a_c= image_a.shape

      # plt.figure()
      # plt.imshow(image_a)
      # plt.show()

      # print(iou)

      # plt.figure()
      # plt.imshow(image_b)
      # plt.show()
      # print()

      image_a = (image_a/255).astype(np.float32)
      image_b = (image_b/255).astype(np.float32)

      rmse_value = np.sqrt(np.sum((image_a-image_b)**2)/a_r/a_c)

      # print(rmse_value)

      compare_RMSE_explanation_list.extend([rmse_value])

  except:
    pass
  print(j)

print(compare_IOU_explanation_list)
print(compare_RMSE_explanation_list)

mean_IOU = np.mean(compare_IOU_explanation_list)


mean_RMSE = np.mean(compare_RMSE_explanation_list)

print(mean_IOU)
print(mean_RMSE)
```
