```python
## load package
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from keras.models import Sequential
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.applications import EfficientNetV2B0 as Model

model = Model(weights='imagenet', include_top=True)
model.compile(
        loss=tf.keras.losses.CategoricalCrossentropy, optimizer=keras.optimizers.Adam(lr=0.01), metrics=["accuracy"]
    )
model.summary()
```

```python
start_idx =0
end_idx = 41

list_of_dir =list()
for i in range(start_idx,end_idx):
  list_of_dir.append(str(i))


start_idx =81
end_idx = 121

for i in range(start_idx,end_idx):
  list_of_dir.append(str(i))
```

```python
result_labels_list = list()
original_labels_list = list()

for j in range(len(list_of_dir)):
# for j in range(5):
  new_files_path = image_path+'/'+str(list_of_dir[j])+'/'
  try:
    dataset = tf.keras.preprocessing.image_dataset_from_directory(new_files_path,
                                                                labels= None,
                                                                label_mode=None,
                                                                image_size=(224, 224),batch_size=32, shuffle=True)
    result = model.predict(dataset)
    result_labels = np.argmax(result, axis=1)
    original_labels=(np.ones(len(result_labels))*int(list_of_dir[j])).astype(np.uint16)

    result_labels_list.extend(result_labels)
    original_labels_list.extend(original_labels)
  except:
    pass

print(result_labels_list)
print(original_labels_list)

accuracy_image = sum(np.array(result_labels_list) == np.array(original_labels_list))/len(result_labels_list)
print(accuracy_image)
```
